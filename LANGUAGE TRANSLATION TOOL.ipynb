```python
import googletrans
from googletrans import Translator

def translate_text(text: str, src_language: str, dest_language: str) -> str:
    translator = Translator()
    try:
        translated = translator.translate(text, src=src_language, dest=dest_language)
        return translated.text
    except Exception as e:
        return f"Error: {str(e)}"

def main():
    print("Simple Language Translation Tool")
    text = input("Enter text to translate: ")
    src_language = input("Enter source language (e.g., 'en' for English): ")
    dest_language = input("Enter target language (e.g., 'es' for Spanish): ")

    translated_text = translate_text(text, src_language, dest_language)
    print(f"Translated text: {translated_text}")

if __name__ == "__main__":
    main()
```
